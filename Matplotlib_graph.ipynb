```python
import matplotlib.pyplot as plt
import numpy as np
```

# Bar Plot

```python
x = ['Python', 'C', 'C++', 'Java']
y = [85, 70, 60, 82]
z = [20, 30, 40, 50]

plt.figure(figsize=(5, 4))
width = 0.2
p = np.arange(len(x))
p1 = [j + width for j in p]


plt.xlabel('Language', fontsize=15)
plt.ylabel('No', fontsize=15)
plt.title('Simple Bar Plot', fontsize=15)

plt.bar(p, y, width, color='r', label='Popularity')
plt.bar(p1, z, width, color='y', label='Popularity1')

plt.xticks(p + width / 2, x) 
plt.legend()


plt.show()
```

# Scatter Plot

```python
day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]

plt.figure(figsize=(5, 4))
colors = [10, 49, 30, 29, 56, 20, 30]
plt.scatter(day, no, c= colors, cmap='viridis')
plt.colorbar()

plt.xlabel('Day', fontsize=12)
plt.ylabel('No', fontsize=12)
plt.title('Scatter Plot', fontsize=14)


plt.show()
```

# Histogram Plot

```python
data = np.random.randn(2000)
plt.figure(figsize=(5, 4))

plt.hist(data, bins= 15, color='green')
```

# Pie Chart / Plot

```python
x = [10, 20, 30, 40]
y = ['C', 'C++', 'java', 'Python']
ex = [0.3, 0.0, 0.0, 0.0]
c = ['r', 'b', 'g', 'y']
plt.pie(x, labels = y, explode = ex, autopct = '%0.1f%%', shadow = True, labeldistance = 1.1, textprops = {"fontsize": 15}
)
plt.title("Distribution of Programing", loc='center')
plt.legend(loc=2)
plt.show()
```

# Stem Plot

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 6, 4, 3]
plt.figure(figsize=(5, 4))
plt.stem(x,y, linefmt=':', markerfmt='o',basefmt='g')
plt.show()
```

# Line Plot

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 15]
plt.figure(figsize=(5, 4))
plt.plot(x, y)
plt.scatter(x, y)
plt.show()
```

# Area/Stack Plot

```python
x = [1, 2, 3, 4, 5]
area1 = [2, 3, 2, 5, 4]
area2 = [2, 3, 4, 5, 6]
area3 = [1, 3, 2, 4, 2]
l = ['area1', 'area2', 'area3']
plt.figure(figsize=(5, 4))
plt.stackplot(x, area1, area2, area3, labels =l)

plt.legend(loc=2)
plt.show()
```

# Box Plot

```python
data = np.array([1, 2, 3, 3, 5, 7, 7, 8, 9, 10, 20])
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1
plt.figure(figsize=(5, 4))

upper_whisker = q3 + (1.5 * iqr)
lower_whisker = q1 - (1.5 * iqr)

plt.boxplot(data)
plt.title('Boxplot Example')
plt.xlabel('Data')
plt.ylabel('Values')

plt.show()
```

# Step Plot

```python
x = [1, 2, 3, 4, 5]
y = [2, 2, 3, 5, 4]
plt.step(x, y, color='r', marker='o', ms=6, mfc='g')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Step Plot Example')
plt.grid()
plt.show()
```

# Fill_Between Plot

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 2, 3, 4, 5])
plt.figure(figsize=(5, 4))
plt.plot(x, y, color='r')
plt.fill_between(x, y, color='green', where=(x>=2) & (x <=4), alpha = 0.7)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Fill_Between Plot')
plt.show()
```
